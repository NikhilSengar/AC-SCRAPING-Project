```python
driver = webdriver.Chrome()
```

```python
data_list = []

for page_url in allPages:
    driver.get(page_url)
    time.sleep(3)  
    
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    acData = soup.find_all('div', class_="tUxRFH")

    for i in acData:
        name = i.find('div', class_="KzDlHZ")
        Ratings = i.find('div', class_="XQDdHH")
        price = i.find('div', class_="Nx9bqj _4b5DiR")
        original_price = i.find('div', class_="yRaY8j ZYYwLA")
        discount_element = i.find('div', class_="UkUFwK")
        features = i.find_all('li')

        name_text = name.text if name else 'N/A'
        ratings_text = Ratings.text if Ratings else 'N/A'
        price_text = price.text[1:] if price else 'N/A'
        original_price_text = original_price.text[1:] if original_price else 'N/A'
        discount_element_text = discount_element.text.split()[0] if discount_element else 'N/A'
        features_text = [feature.text for feature in features] if features else []

        dict_ = {
            'Name': name_text,
            'Ratings': ratings_text,
            'Price': price_text,
            'Original Price': original_price_text,
            'Discount Element': discount_element_text,
            'Features': features_text
        }
        data_list.append(dict_)
```
